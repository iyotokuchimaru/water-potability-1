```python
import pandas as pd
import numpy as np

data = pd.read_csv('https://raw.githubusercontent.com/DamonGreenhalgh/water-potability/main/water_potability.csv')

data.head()
```

```python
def train_test_set(X):
  new_X=X.sample(frac=1)
  separate=int(len(new_X.index)*0.8)
  train=new_X.iloc[:separate]
  test=new_X.iloc[separate:]
  return[train,test]
```

```python
[train_data,test_data]=train_test_set(data)
```
